```
options={
    "header":True,
    "inferschema":True,
    "delimeter":","
}
def read_csv(format_name,options,path):
    return spark.read.format(format_name).options(**options).load(path)

df=read_csv('csv',options,'dbfs:/mnt/adfassign/customer/20240105_sales_customer.csv')
```

```
display(df)
```

```
#1st
#•	All the column header should be in snake case in lower case (By using UDF function dynamically works for all camel case to snake case)

def snake_case(i):
    x=i.lower()
    return x.replace(' ','_')
```

```
x=udf(snake_case,StringType())

lst=list(map(lambda i:snake_case(i),df.columns))
df1=df.toDF(*lst)
display(df1)
```

```
#2
# •	 By using the "Name" column split by " " and create two columns first_name and last_name

col_split=df1.withColumn('fname',split(col('name'),' ')[0]).withColumn('lname',split(col('name'),' ')[1])
display(col_split)
```

```
#3
#•	Create column domain and extract from email columns Ex: Email = "josephrice131@slingacademy.com" domain="slingacademy"

extract_col=df1.withColumn('domain',split(col('email_id'),'@')[1])
display(extract_col)
```

```
#4
#	Create a column gender where male = "M" and Female="F"
gen_col=df1.withColumn("gender",when(df1.gender == "male","M").when(df1.gender == "female","F").otherwise(df1.gender))
display(gen_col)
```

```
#5
#•	From Joining date create two colums date and time by splitting based on " " delimiter.

joiningdate=df1.withColumn('date',split(col('joining_date'),' ')[0]).withColumn('time',split(col('joining_date'),' ')[1])
display(joiningdate)
```

```
#6
#•	 Date column should be on "yyyy-MM-dd" format.

date_col=joiningdate.withColumn('date',date_format(to_date(col('date'),'dd-MM-yyyy'),'yyyy-MM-dd'))
display(date_col)
```

```
#7
#•	 Create a column expenditure-status, based on spent column is spent below 200 column value is "MINIMUM" else "MAXIMUM"

expenditure_col=df1.withColumn('expenditure-status',when(df1.spent<'200','MINIMUN').otherwise('MAXIMUM'))
display(expenditure_col)
```

```
#8
#•	 Write based on upsert [table_name: customer] (in silver layer path is silver/sales_view/tablename/{delta pearquet}


def saveToDeltaWithOverwrite1(resultDf, silverLayerPath,database_name,target_table_name, mergeCol):
    base_path=silverLayerPath+f"/{target_table_name}"
    mappedCol = " AND ".join(list(map((lambda x: f"old.{x} = new.{x} "),mergeCol)))
    if not DeltaTable.isDeltaTable(spark,f"{base_path}"):
        spark.sql(f"CREATE DATABASE IF NOT EXISTS {database_name}")
        resultDf.write.mode("overwrite").format("delta").option("path",base_path).saveAsTable(f"{database_name}.{target_table_name}")
    else:
        deltaTable = DeltaTable.forPath(spark, f"{base_path}")
        deltaTable.alias("old").merge(resultDf.alias("new"),mappedCol)\
        .whenMatchedUpdateAll()\
        .whenNotMatchedInsertAll()\
        .execute()
```
