```
import requests
 
def make_api_request(api_url, bearer_token):
    headers = {
        'Authorization': f'Bearer {bearer_token}',
        'Content-Type': 'application/json',
    }
 
    response = requests.get(url = api_url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        response.raise_for_status()
 

```

```
import requests
from requests.auth import HTTPBasicAuth

def get_bearer_token(client_email, client_name,token_url):
    # Prepare the token request payload
    data = {
        'client_email': client_email,
        'client_name': client_name
    }

    # Make the token request
    response = requests.post(token_url, data=data)

    if response.status_code == 200:
        # Successfully obtained the token
        token_data = response.json()
        # print('token_data is :',token_data)
        return token_data.get('access_token')
    else:
        print(f"Failed to obtain token. Status code: {response.status_code}")
        print(response.text)
        return None

client_email = 'rest2@gmail.com'
client_name = 'rest2'
token_url = 'https://simple-books-api.glitch.me/books'

bearer_token = get_bearer_token(client_email, client_name, token_url)
if bearer_token:
    print(f"Bearer Token: {bearer_token}")
else:
    print("Failed to obtain Bearer Token.")
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode
from pyspark.sql.types import StructType, StructField, StringType, ArrayType

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("Explode Data") \
    .getOrCreate()

# Define schema for the data
schema = StructType([
    StructField("web_pages", ArrayType(StringType())),
    StructField("name", StringType()),
    StructField("domains", ArrayType(StringType())),
    StructField("country", StringType()),
    StructField("state_province", StringType()),
    StructField("alpha_two_code", StringType())
])

# Sample data
data = [
    (['http://www.aabfs.org/'], 'Arab Academy for Banking and Financial Sciences', ['aabfs.org'], 'Jordan', None, 'JO'),
    (['http://www.aabu.edu.jo/'], 'Al al-Bayt University', ['aabu.edu.jo'], 'Jordan', None, 'JO'),
]

# Create DataFrame
df = spark.createDataFrame(data, schema)

# Explode columns with array type
df_exploded = df.select(explode(df.web_pages).alias("web_pages"), df.name, explode(df.domains).alias("domain"), df.country, df.state_province, df.alpha_two_code)

# Show the exploded DataFrame
df_exploded.show(truncate=False)

# Stop SparkSession
spark.stop()
```
